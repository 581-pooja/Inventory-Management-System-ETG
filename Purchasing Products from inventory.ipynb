# For Generating the bill of the customer transactions , Calculating the sales of the day and showing the products present in the inventory

```python
# Importing packages
import time
import math
import json

# All dictionary and list required for sales recording
sales_dict = {}
trans_id_list = []
    
def generating_bill():

    # initalize the transaction id for soomth funcitoning
    trans_id = 1

    # Taking customer name and how many product customer wants
    n_purchase_prod = int(input("Enter the number of the products that customer want: "))

    for i in range(0,n_purchase_prod):
        us_prod_id = int(input("Enter the product Id: "))

        # if user written wrong product id
        if us_prod_id not in records.keys():
            print("Product does not exists in inventory...")
        else:
            cust_name = input("Enter Customer name:")

            # if product is out of stock
            if(records[us_prod_id]['Quantity'] == 0):
                print("The Product is Out of Stock. Have some other items\n")
                us_prod_id = int(input("Enter the product Id: "))

            us_quan = int(input("Enter Quantity required by customer: "))

            # if customer wants products more than the persent in the inventory    
            if(us_quan > records[us_prod_id]['Quantity']):
                print("The Quantity of product available is:", records[us_prod_id]['Quantity'])
                permis = input("Do you want max product quantity yes or no :")
                if(permis == 'yes'):
                    us_quan = records[us_prod_id]['Quantity']
                else:
                    us_quan = int(input("Enter Quantity required by customer: "))

            # Printing the Bill for customers
            print("******************************")
            print("Inventory Management System")
            print(time.ctime())
            print("Customer Name: ", cust_name)
            print("******************************")
            print("Product Id:", us_prod_id)
            print("Product name:", records[us_prod_id]['name'])
            print("Product Price:", records[us_prod_id]['price'])
            print("Product Quantity:", records[us_prod_id]['Quantity'])
            print("Product Net weight:", records[us_prod_id]['Net weight'])
            print("Product Discount:", records[us_prod_id]['Discount'])
            print("******************************")

            # calculating the price according to the discount
            #discounted_price = original_price - (original_price * discount / 100)
            original_price = int(records[us_prod_id]['price'])

            discount = records[us_prod_id]['Discount']

            # Calculating the discount on the products
            if(discount != 0):
                discounted_price =  original_price - math.floor((original_price * discount) / 100)
                amount_pay = math.ceil(discounted_price * us_quan)
            else:
                amount_pay = math.ceil(original_price * us_quan)

            print("Billing Amount: ", amount_pay)
            print("******************************")

            # appending the purchase of the product in sales_dict
            sales_dict[trans_id] = {
                'Product_id': us_prod_id,
                'Customer_name' : cust_name,
                'product_name': records[us_prod_id]['name'],
                'product_price': amount_pay,
                'product_quantity': records[us_prod_id]['Quantity'],
                'product_discount': records[us_prod_id]['Discount'],
            }

            # all updates for sales recording
            trans_id_list.append(trans_id)
            trans_id = trans_id + 1

            # update the quantity in records.json after purchasing the items
            final_quantity = math.floor(records[us_prod_id]['Quantity'] - us_quan)
            records[us_prod_id]['Quantity'] = final_quantity
    
    # After purchasing the product writing it in json file : Update Inventory after Purchase
    js_update = json.dumps(records)
    fd = open("records.json","w")
    fd.write(js_update)
    fd.close()

# For showing products present in the inventory
def showing_products():
    print(f"Product Id:\t Product Name:\t\t Price:\t\t Quantity:\t Net weight:\t Discount:")
    print("----------------------------------------------------------------------------------------------------------------------")
    for prod_id in records.keys():
        print(f"{prod_id}\t\t {records[prod_id]['name'].rjust(10)}\t\t {(str(records[prod_id]['price'])).rjust(6)}\t\t {str(records[prod_id]['Quantity']).rjust(5)}\t\t {records[prod_id]['Net weight'].rjust(6)}\t\t {str(records[prod_id]['Discount']).rjust(6)}")
    print("----------------------------------------------------------------------------------------------------------------------")

# For generating the sales.txt for everyday transactions
def generating_sales():
    # Writing the heading initially
    fd = open("sales.txt","w")
    fd.write(f"Date of the Transactions is: {time.ctime()}\n")
    fd.write(f"Transacion Id: Product Id: Product Name: Customer Name: Quantity: Amount Paid \n")
    fd.write(f"**************************************************************\n")
    fd.close()

    # printing the products sales details
    total_amount = 0
    for trans_id in trans_id_list:
        if trans_id in sales_dict.keys(): 
            fd = open("sales.txt","a")
            fd.write(f"\n{trans_id}\t\t{sales_dict[trans_id]['Product_id']}\t\t {sales_dict[trans_id]['product_name']}\t\t {sales_dict[trans_id]['Customer_name']}\t\t {sales_dict[trans_id]['product_quantity']}\t\t {sales_dict[trans_id]['product_price']}\n")
            total_amount = total_amount + sales_dict[trans_id]['product_price']
            fd.close()

    # for calculating the total price of the day
    f = open("sales.txt","a")
    f.write(f"**************************************************************\nTotal Sales of the day is: {total_amount}\n**************************************************************\n")
    f.close()
    print("Your Transactions are added in Sales.txt .. Thankyou!")
```

# Main

```python
print("\t-- Welcome to Inventory Management System --")
while (1):
    print("\n*****************************************************************")
    print("\t 1 for Printing products from Inventory \n\t 2 for purchasing and Greating the bill of Purchased products \n\t 3 Greating sales.txt transaction of the day \n\t 4 for Quiting from the Inventory")
    print("*****************************************************************\n")
    option = int(input("Enter your option : "))
    
    if option == 1:
        showing_products()
    elif option == 2:
        generating_bill()
    elif option == 3:
        generating_sales()
    elif option == 4:
        print("*****************************************************************")
        print("\tThankyou !! Have a great Day!!")
        print("*****************************************************************")
        break
    else:
        print("Please put valid input")
```
