# For adding , deleting , modifying and Showing the products present in the Inventory

```python
import json
        
# To add items from inventory
def add_items():
    n_times = int(input("Enter how many product you want to add to the inventory:"))
    for i in range(1,n_times+1):
        prod_id = int(input("\nEnter Product id:"))
        
        if prod_id in records.keys():
            ans = input("do you want to modify the items present in inventory: Enter 1 or yes and 0 or no:")
            if(ans == '1' or ans == "yes"):
                print("Product will be Modified!")
            else:
                print("Product is persent with above id. Products cannot exists with duplicate id.")
                break
        
        prod_name = input("Enter Product Name: ")
        prod_price = int(input("Enter Product Price: "))
        prod_quan = int(input("Enter Product Quantity: "))
        prod_weight = input("Enter Product Net weight: ")
        prod_discount = int(input("Enter Product Discount: "))
        records[prod_id] = {'name': prod_name, 'price':prod_price, 'Quantity': prod_quan, 'Net weight': prod_weight,'Discount': prod_discount}
    
    # adding the products in the json file
    js = json.dumps(records)
    fd = open("records.json",'w') 
    fd.write(js)
    fd.close()
    
    if(n_times>0):
        print("Your Products are added Successfully!")
    else:
        print("NO Products added to Inventory!")

# To delete items from inventory
def delete_items():
    n_times = int(input("Enter how many product you want to delete from the inventory:"))
    for i in range(1,n_times+1):
        prod_id = int(input("\nEnter Product id:"))
        if prod_id in records.keys():
            del records[prod_id]
            print("Product Successfully deleted from Inventory!")
        else:
            print("Product does not exist in inventory.")
            
    # making changes in inventory
    js = json.dumps(records)
    fd = open("records.json",'w') 
    fd.write(js)
    fd.close()

# To modify items preseny in the inventory
def modify_items():
    ans = input("Do you want to modify all then enter all or 1 \nOnly price then enter pr or 2\nonly Quantity then enter qn or 3 :")
    if(ans == 'all' or ans == '1'):
        # To modify all the features of the product
        n_times = int(input("Enter how many products you want to modify in the inventory:"))
        for i in range(1,n_times+1):
            prod_id = int(input("\nEnter Product id:"))
            if prod_id in records.keys():
                prod_name = input("Enter Product Name: ")
                prod_price = int(input("Enter Product Price: "))
                prod_quan = int(input("Enter Product Quantity: "))
                prod_weight = input("Enter Product Net weight: ")
                prod_discount = int(input("Enter Product Discount: "))
                records[prod_id] = {'name': prod_name, 'price':prod_price, 'Quantity': prod_quan, 'Net weight': prod_weight,'Discount': prod_discount}
                print("Product Successfully Modified from Inventory!")
            else:
                print("Product does not exist in inventory.")
                
    elif(ans == 'pr' or ans == '2'):
        n_times = int(input("Enter how many products price you want to modify in the inventory:"))
        for i in range(1,n_times+1):
            prod_id = int(input("Enter Product id:"))
            if prod_id in records.keys():
                prod_price = int(input("Enter Product Price: "))
                records[prod_id]['price'] = prod_price
                print("Product Successfully Modified from Inventory!")
            else:
                print("Product does not exist in inventory.")
                
    elif(ans == 'qn' or ans == '3'):
        # To modify only quantity of the products as there can be new products added too.
        n_times = int(input("Enter how many products quantity you want to modify in the inventory:"))
        for i in range(1,n_times+1):
            prod_id = int(input("Enter Product id:"))
            if prod_id in records.keys():
                prod_quan = int(input("Enter Product Quantity to modify: "))
                records[prod_id]['Quantity'] = prod_quan
                print("Product Successfully Modified from Inventory!")
            else:
                print("Product does not exist in inventory.")
    else:
        print("Invalid Input.")

    # making changes in inventory
    js = json.dumps(records)
    fd = open("records.json",'w') 
    fd.write(js)
    fd.close()
        
#  to print items present in the inventory
def print_records():
    ans = input("Do you want to see: All records then enter 1 or all : For specific records enter 0 or sp:")
    if(ans == 'all' or ans == '1'):
        print(f"Product Id:\t Product Name:\t\t Price:\t\t Quantity:\t Net weight:\t Discount: \n")
        print("----------------------------------------------------------------------------------------------------------------------")
        for prod_id in records.keys():
            print(f"{prod_id}\t\t {records[prod_id]['name'].rjust(10)}\t\t {(str(records[prod_id]['price'])).rjust(6)}\t\t {str(records[prod_id]['Quantity']).rjust(5)}\t\t {records[prod_id]['Net weight'].rjust(6)}\t\t {str(records[prod_id]['Discount']).rjust(6)}")
        print("----------------------------------------------------------------------------------------------------------------------")
    
    elif(ans == 'sp' or ans == '0'):
        prod_id = int(input("Enter the Product id to view specific product:"))
        print(f"Product Id:\t Product Name:\t\t Price:\t\t Quantity:\t Net weight:\t Discount: ")
        print("-------------------------------------------------------------------------------------------")
        if prod_id in records.keys():
            print(f"{prod_id}\t\t {records[prod_id]['name']}\t\t {records[prod_id]['price']}\t\t {records[prod_id]['Quantity']}\t\t {records[prod_id]['Net weight']}\t\t {records[prod_id]['Discount']}\t\t  ")
            print("-------------------------------------------------------------------------------------------")
        else:
            print("Product does not exist in inventory.")
    else:
        print("Invalid Input.")
        
```
